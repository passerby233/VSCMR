# Compare stoies generated by different methods

```python
import json
import os.path
import sys
from PIL import Image
from matplotlib import pyplot as plt

sys.path.append('../')
from rule_mining import utils as ut
from rule_mining import config as cfg

XE_pred_file = "/home/lijiacheng/AREL-master/data/save/XE/XE_b3.json"
seman_XE_file = '/home/lijiacheng/VSCMR/data/save/XE2_2/XE2-2_b1.json'
semantic_list_file = os.path.join(cfg.proj_dir, 'VIST/semantic_list.json')
image_dir = os.path.join(cfg.dataset_path, 'images/test')

XE_stories = ut.load_json(XE_pred_file)['output_stories']
seman_XE_stories = ut.load_json(seman_XE_file)['output_stories']

semantic_list = ut.load_json(semantic_list_file)
id2words = ut.load_json(cfg.txtdata_file)['id2words']
story_line = ut.load_json(cfg.story_line_file)
```

```python
def convert_to_dict(stories):
    story_dict = {}
    for story in stories:
        story_dict[story['story_id']] = story
    stories = story_dict
    return stories
XE_stories = convert_to_dict(XE_stories)
seman_XE_stories = convert_to_dict(seman_XE_stories)

def show_single_story(story_id):

    semantic = [[id2words[str(idx-2048)] for idx in img_list] 
                for img_list in semantic_list['test'][story_id]]
    plt.figure(figsize=(17, 90))
    for i in range(5):
        base_name = seman_XE_stories[story_id]['photo_sequence'][i] + ".jpg"
        filename = os.path.join(image_dir, base_name)
        if not os.path.exists(filename):
            base_name = seman_XE_stories[story_id]['photo_sequence'][i] + ".png"
            filename = os.path.join(image_dir, base_name)
        image = Image.open(filename)
        plt.subplot(1, 5, 1+i)
        plt.imshow(image)
        plt.axis('off') 
    plt.show()
    print('story_id: %s' % story_id)
    print('photo_flickrs: ', seman_XE_stories[img_id]['photo_sequence'])
    print("--XE--:")
    print " ",XE_stories[story_id]['story_text_normalized'],'\n'
    print("--seman_XE--:")
    print " ",seman_XE_stories[story_id]['story_text_normalized'],'\n'
    print("--Ground_truth--:")
    print "  ",story_line['test'][story_id]['origin_text'],'\n'
    print"--semantic:",semantic
```

#### To show storiy comparison in a range

```python
for story_id in seman_XE_stories.keys()[400:600]:
    if story_id in XE_stories and story_id in seman_XE_stories:
        show_single_story(story_id)
```

#### To show single story comparison

```python
story_id = '50221'
show_single_story(story_id)
```
